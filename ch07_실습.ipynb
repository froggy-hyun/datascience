```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-10, 10, 0.1)
f_x = x**2 - 4*x

plt.plot(x, f_x, label="C(w0) = w0^2 - 4w0")
plt.title("Function C(w0) = w0^2 - 4w0")
plt.xlabel("w0")
plt.ylabel("C(w0)")
plt.grid()
plt.legend()
plt.show()

w0 = 1 
learning_rate = 0.1
w0_list = []
c_list = []

for i in range(1000): 
    gradient = 2*w0 - 4
    w0 = w0 - learning_rate * gradient
    w0_list.append(w0)
    c_list.append(w0**2 - 4*w0)
    
    if abs(gradient) < 1e-6:
        break

plt.plot(x, f_x, label="C(w0) = w0^2 - 4w0")
plt.scatter(w0_list, c_list, color='red', label="Gradient Descent Path")
plt.title("Gradient Descent on C(w0) = w0^2 - 4w0")
plt.xlabel("w0")
plt.ylabel("C(w0)")
plt.grid()
plt.legend()
plt.show()

print(f"최적 w0 값: {w0}")
print(f"최소 C(w0) 값: {w0**2 - 4*w0}")
```

```python
x = [1, 2, 3, 4]
y = [2, 5, 5, 9]

def J(w1):
    sum = 0
    for i in range(len(x)):
        f = (x[i] * w1 - y[i]) ** 2
        sum += f
    return sum

def J_grad(w1):
    grad = 0
    for i in range(len(x)):
        grad += 2 * (x[i] * w1 - y[i]) * x[i]
    return grad

w1 = 1 
learning_rate = 0.001
epochs = 1000

for i in range(epochs):
    grad = J_grad(w1)
    w1 = w1 - learning_rate * grad

print('w1 =', w1)
print('C(w1) =', J(w1))
```

```python
import numpy as np

X = np.array([1, 2, 3, 4]) 
y = np.array([2, 5, 5, 8]) 
m = len(X) 

w0 = 1.0
w1 = 1.0

lr = 0.001
epochs = 1000

def jgrad(w0, w1, X, y):
    gw0 = 0
    gw1 = 0
    for i in range(len(X)):
        error = w1 * X[i] + w0 - y[i]
        gw0 += (1 / m) * error
        gw1 += (1 / m) * error * X[i]
    return gw0, gw1

def J(w0, w1, X, y):
    result = 0
    for i in range(len(X)):
        result += (1 / (2 * m)) * (w1 * X[i] + w0 - y[i])**2
    return result

for epoch in range(epochs):
    grad_w0, grad_w1 = jgrad(w0, w1, X, y)
    
    w0 -= lr * grad_w0
    w1 -= lr * grad_w1
    
    if epoch % 100 == 0:
        print(f"Epoch {epoch}: Cost = {J(w0, w1, X, y):.4f}, w0 = {w0:.4f}, w1 = {w1:.4f}")

print("\n최종 결과:")
print(f"w0 = {w0:.6f}")
print(f"w1 = {w1:.6f}")
print(f"최종 함수값 J = {J(w0, w1, X, y):.6f}")
```
